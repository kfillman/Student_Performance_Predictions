# Student Performance data

## Imports

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import data
file_path = "Student_performance_data.csv"  # Ensure this file is in the same directory
df = pd.read_csv(file_path)

df.head()
```

## EDA

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics
display(df.describe())
```

```python
df.dtypes
```

Dataset has no missing values and is entirely numeric.

## Visualizing Feature Distribution

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_cols].hist(figsize=(12, 8), bins=20)
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

# Correlation Analysis

```python
# Make matrix for numeric features
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
